Q11. Ask user to enter the marks and define if the user got distinction, first class, second class, pass, Fails

If marks are more than 80 output must be “You got distinction”

For marks more than 60 output must be “You got First class”

Marks more than 40 will display “You got second class”

Marks more than or equal to 35 “PASS” Otherwise Fail

Basic if-else version

```python
marks = int(input("Enter the marks: "))

if marks > 80:
    print("You got distinction")
elif marks > 60:
    print("You got First class")
elif marks > 40:
    print("You got second class")
elif marks >= 35:
    print("PASS")
else:
    print("Fail")
```

Function version

```python
import sys
def grade(marks):
    if marks > 80:
        output = "You got Distinction"
    elif marks > 60:
        output = "You got First Class"
    elif marks > 40:
        output = "You got Second Class"
    elif marks >= 35:
        output = "PASS"
    else:
        output = "Fail"

    return output


while True:
    try:
        marks = int(input("Enter the marks (0–100): "))

        if 0 <= marks <= 100:
            print(grade(marks))
            break
        else:
            print("Marks must be b etween 0 and 100.Please Try again!!!")

    except Exception as e:
        print("Invalid input. Please enter a valid integer!!!",sys.exc_info,e)
```

OOPS version

```python
import sys

class Grade:

    def __init__(self, marks):
        self.marks = marks

    def calculate(self):
        if self.marks > 80:
            return "You got Distinction"
        elif self.marks > 60:
            return "You got First Class"
        elif self.marks > 40:
            return "You got Second Class"
        elif self.marks >= 35:
            return "PASS"
        else:
            return "Fail"


while True:
    try:
        marks = int(input("Enter the marks (0–100): "))

        if 0 <= marks <= 100:
            student = Grade(marks)
            print(student.calculate())
            break
        else:
            print("Marks must be between 0 and 100. Please try again!")

    except ValueError:
        print("Invalid input. Please enter a valid integer!!!")
    except Exception:
        print("Something went wrong:", sys.exc_info())
```

SQL lite version

```python
import sqlite3

class Grade:

    def __init__(self, marks):
        self.marks = marks

    def calculate(self):
        if self.marks > 80:
            return "You got Distinction"
        elif self.marks > 60:
            return "You got First Class"
        elif self.marks > 40:
            return "You got Second Class"
        elif self.marks >= 35:
            return "PASS"
        else:
            return "Fail"


conn = sqlite3.connect("grades.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS student_grade (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    marks INTEGER,
    grade TEXT
)
""")

while True:
    try:
        marks = int(input("Enter the marks (0–100): "))

        if 0 <= marks <= 100:
            student = Grade(marks)
            result = student.calculate()

            cursor.execute(
                "INSERT INTO student_grade (marks, grade) VALUES (?, ?)",
                (marks, result)
            )
            conn.commit()

            print(result)
            print("Record saved successfully!")
            break
        else:
            print("Marks must be between 0 and 100. Please try again!")

    except ValueError:
        print("Invalid input. Please enter a valid integer!!!")

conn.close()
```

```python
import sqlite3

class SalaryIncrement:

    def __init__(self, salary, rating):
        self.salary = salary
        self.rating = rating.upper()
        self.salary_amount = salary * 100000
        self.increment_percentage = 0
        self.new_salary = 0

    def calculate_increment(self):

        if self.salary <= 5:
            rates = {'A':0.16, 'B':0.12, 'C':0.10, 'D':0.06}
        elif self.salary <= 10:
            rates = {'A':0.14, 'B':0.10, 'C':0.08, 'D':0.06}
        elif self.salary <= 15:
            rates = {'A':0.08, 'B':0.06, 'C':0.04, 'D':0.00}
        elif self.salary <= 23:
            rates = {'A':0.07, 'B':0.05, 'C':0.04, 'D':0.00}
        else:
            rates = {}

        self.increment_percentage = rates.get(self.rating, 0)
        increment_amount = self.salary_amount * self.increment_percentage
        self.new_salary = self.salary_amount + increment_amount

    def save_to_db(self, cursor):
        cursor.execute("""
        INSERT INTO employee_salary
        (salary_lpa, rating, increment_percent, new_salary)
        VALUES (?, ?, ?, ?)
        """, (self.salary, self.rating,
              self.increment_percentage * 100,
              self.new_salary))

    def display(self):
        print(f"Original Salary: {self.salary_amount:.2f}")
        print(f"Rating: {self.rating}")
        print(f"Increment %: {self.increment_percentage * 100:.0f}%")
        print(f"New Salary: {self.new_salary:.2f}")



conn = sqlite3.connect("salary.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS employee_salary (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    salary_lpa REAL,
    rating TEXT,
    increment_percent REAL,
    new_salary REAL
)
""")

while True:
    print("\n1. Add Employee")
    print("2. Display Records")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        try:
            salary = float(input("Enter salary (LPA): "))
            rating = input("Enter rating (A/B/C/D): ").upper()

            if salary <= 0 or salary > 23:
                print("Salary must be between 0 and 23 LPA.")
                continue

            if rating not in ['A','B','C','D']:
                print("Invalid rating.")
                continue

            emp = SalaryIncrement(salary, rating)
            emp.calculate_increment()
            emp.save_to_db(cursor)
            conn.commit()

            emp.display()
            print("Record saved successfully!")

        except ValueError:
            print("Enter numeric salary only.")

    elif choice == "2":
        cursor.execute("SELECT * FROM employee_salary")
        records = cursor.fetchall()

        if records:
            print("\n--- Employee Records ---")
            for row in records:
                print(f"ID:{row[0]} | Salary(LPA):{row[1]} | "
                      f"Rating:{row[2]} | Increment%:{row[3]} | "
                      f"New Salary:{row[4]:.2f}")
        else:
            print("No records found.")

    elif choice == "3":
        break

    else:
        print("Invalid choice.")

conn.close()
```

Q13 . Ask user to opt for courses for master degree based on the following

L1 = [“HR”, “Finance”, “Marketing”, “DS”]

Based on above subject there are two different streams. For example- HR is having HR core and HR analytics and Marketing is having core and Marketing analytics. Analytics is the optional subject and having added extra fees. DS is not having analytics.

If fees for L1 is 2 lakhs for each course core subject having the same fees but analytics subject having 10% extra on 2 lakhs.

If student opts for hostel then 2 lakhs per year is added. For food monthly 2000 .

Transportation charges 13000 per semester. Calculate the total annual cost based on selected service.

User will enter values as subject, analytics(Y/N), Hostel (Y/N), food(How many months?), Transportation(semester/annual)

if-else version

```python
L1 = ["HR", "Finance", "Marketing", "DS"]

BASE_FEE = 200000
ANALYTICS_EXTRA = 0.10
HOSTEL_FEE = 200000
FOOD_PER_MONTH = 2000
TRANSPORT_PER_SEM = 13000


subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
analytics = input("Analytics? (Y/N): ").strip().upper()
hostel = input("Hostel? (Y/N): ").strip().upper()
food_months = int(input("Food for how many months?: "))
transport = input("Transportation (semester/annual): ").strip().lower()


if subject not in L1:
    print("Invalid subject selected!")
    exit()


total_fee = BASE_FEE


if analytics == "Y" and subject != "DS":
    total_fee += BASE_FEE * ANALYTICS_EXTRA


if hostel == "Y":
    total_fee += HOSTEL_FEE

total_fee += food_months * FOOD_PER_MONTH

if transport == "semester":
    total_fee += TRANSPORT_PER_SEM * 2
elif transport == "annual":
    total_fee += TRANSPORT_PER_SEM * 2
else:
    print("Invalid transportation option!")
    exit()

print("\n----- Fee Breakdown -----")
print(f"Subject: {subject}")
print(f"Total Annual Cost: ₹{int(total_fee)}")
```

Function version

```python
L1 = ["HR", "Finance", "Marketing", "DS"]

BASE_FEE = 200000
ANALYTICS_EXTRA = 0.10
HOSTEL_FEE = 200000
FOOD_PER_MONTH = 2000
TRANSPORT_PER_SEM = 13000

def calculate_total_fee():

    while True:
        subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
        if subject in L1:
            break
        print("Invalid subject. Please choose from HR, Finance, Marketing, DS.")

    while True:
        analytics = input("Analytics? (Y/N): ").strip().upper()
        if analytics in ["Y", "N"]:
            break
        print("Enter Y or N only.")

    while True:
        hostel = input("Hostel? (Y/N): ").strip().upper()
        if hostel in ["Y", "N"]:
            break
        print("Enter Y or N only.")

    while True:
        try:
            food_months = int(input("Food for how many months?: "))
            if food_months >= 0:
                break
            print("Months cannot be negative.")
        except ValueError:
            print("Please enter a valid number.")

    while True:
        transport = input("Transportation (semester/annual): ").strip().lower()
        if transport in ["semester", "annual"]:
            break
        print("Enter 'semester' or 'annual' only.")

    total_fee = BASE_FEE

    if analytics == "Y" and subject != "DS":
        total_fee += BASE_FEE * ANALYTICS_EXTRA

    if hostel == "Y":
        total_fee += HOSTEL_FEE

    total_fee += food_months * FOOD_PER_MONTH

    total_fee += TRANSPORT_PER_SEM * 2

    print("\n----- Fee Breakdown -----")
    print(f"Subject: {subject}")
    print(f"Total Annual Cost: ₹{int(total_fee)}")


calculate_total_fee()
```

```python
import sqlite3

class FeeCalculator:

    L1 = ["HR", "Finance", "Marketing", "DS"]

    BASE_FEE = 200000
    ANALYTICS_EXTRA = 0.10
    HOSTEL_FEE = 200000
    FOOD_PER_MONTH = 2000
    TRANSPORT_PER_SEM = 13000

    def __init__(self, subject, analytics, hostel, food_months, transport):
        self.subject = subject
        self.analytics = analytics
        self.hostel = hostel
        self.food_months = food_months
        self.transport = transport
        self.total_fee = 0

    def calculate_fee(self):

        self.total_fee = self.BASE_FEE

        if self.analytics == "Y" and self.subject != "DS":
            self.total_fee += self.BASE_FEE * self.ANALYTICS_EXTRA

        if self.hostel == "Y":
            self.total_fee += self.HOSTEL_FEE

        self.total_fee += self.food_months * self.FOOD_PER_MONTH

        self.total_fee += self.TRANSPORT_PER_SEM * 2

    def save_to_db(self, cursor):
        cursor.execute("""
        INSERT INTO student_fee
        (subject, analytics, hostel, food_months, transport, total_fee)
        VALUES (?, ?, ?, ?, ?, ?)
        """, (self.subject, self.analytics, self.hostel,
              self.food_months, self.transport, self.total_fee))

    def display(self):
        print("\n----- Fee Breakdown -----")
        print(f"Subject: {self.subject}")
        print(f"Total Annual Cost: ₹{int(self.total_fee)}")



conn = sqlite3.connect("fees.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS student_fee (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    subject TEXT,
    analytics TEXT,
    hostel TEXT,
    food_months INTEGER,
    transport TEXT,
    total_fee REAL
)
""")

while True:
    print("\n1. Add Student Fee")
    print("2. Display Records")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        while True:
            subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
            if subject in FeeCalculator.L1:
                break
            print("Invalid subject.")

        analytics = input("Analytics? (Y/N): ").strip().upper()
        hostel = input("Hostel? (Y/N): ").strip().upper()
        food_months = int(input("Food months: "))
        transport = input("Transportation (semester/annual): ").strip().lower()

        student = FeeCalculator(subject, analytics, hostel, food_months, transport)
        student.calculate_fee()
        student.save_to_db(cursor)
        conn.commit()

        student.display()
        print("Record saved successfully!")

    elif choice == "2":
        cursor.execute("SELECT * FROM student_fee")
        records = cursor.fetchall()

        if records:
            print("\n--- Student Fee Records ---")
            for row in records:
                print(f"ID:{row[0]} | Subject:{row[1]} | "
                      f"Total Fee: ₹{int(row[6])}")
        else:
            print("No records found.")

    elif choice == "3":
        break

    else:
        print("Invalid choice.")

conn.close()
```

Q44. Create a table  cust_info as sr_no, name, DOB, mobile. Ask user to enter the information from python code. Validate all fields and after validation insert records in the table.

```python
import sqlite3
import re

conn = sqlite3.connect("customer.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS cust_info (
    sr_no INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    DOB TEXT NOT NULL,
    mobile TEXT NOT NULL
)
""")

while True:
    name = input("Enter Name: ").strip()
    if not name.replace(" ", "").isalpha():
        print("Invalid name. Only alphabets allowed.")
        continue

    dob = input("Enter DOB (DD-MM-YYYY): ").strip()
    if not re.match(r"\d{2}-\d{2}-\d{4}$", dob):
        print("Invalid DOB format.")
        continue

    mobile = input("Enter Mobile Number: ").strip()
    if not (mobile.isdigit() and len(mobile) == 10):
        print("Invalid mobile number.")
        continue

    cursor.execute(
        "INSERT INTO cust_info (name, DOB, mobile) VALUES (?, ?, ?)",
        (name, dob, mobile)
    )
    conn.commit()

    print("Record inserted successfully!")

    choice = input("Add another record? (yes/no): ")
    if choice.lower() != "yes":
        break


print("\n----- Customer Records -----")
cursor.execute("SELECT * FROM cust_info")
records = cursor.fetchall()

for row in records:
    print(f"SR_NO: {row[0]} | Name: {row[1]} | DOB: {row[2]} | Mobile: {row[3]}")

conn.close()
```

Q47. Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.

```python
from datetime import date

def calculate_age(dob):
    today = date.today()
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))

def check_eligibility(dob, eligibility_age):
    age = calculate_age(dob)
    return age >= eligibility_age, age

while True:
    try:
        y = input("Enter birth year: ").strip()
        m = input("Enter birth month: ").strip()
        d = input("Enter birth day: ").strip()
        e_age = input("Enter eligibility age: ").strip()

        if not (y.isdigit() and m.isdigit() and d.isdigit() and e_age.isdigit()):
            print("Enter only numeric values\n")
            continue

        y, m, d, e_age = int(y), int(m), int(d), int(e_age)
        today = date.today()

        if y < 1900 or y > today.year:
            print("Invalid year\n")
            continue

        if m < 1 or m > 12:
            print("Invalid month\n")
            continue

        if e_age <= 0 or e_age > 120:
            print("Invalid eligibility age\n")
            continue

        try:
            dob = date(y, m, d)
        except ValueError:
            print("Invalid day for given month\n")
            continue

        if dob > today:
            print("DOB cannot be in future\n")
            continue

        eligible, age = check_eligibility(dob, e_age)

        print("Customer Age:", age)
        if eligible:
            print("Customer is ELIGIBLE")
        else:
            print("Customer is NOT ELIGIBLE")

        break

    except Exception:
        print("Unexpected error. Try again.\n")
```

Q48 .Create a function to check if string is palindrome or not ? For example, if input is NITIN then reverse of the string is same then it is palindrome. If input is ANIL then reverse is LINA which is not same then it is not palindrome.

```python
def is_palindrome(s):
    try:
        if not isinstance(s, str):
            raise ValueError("Input must be a string")
        s = s.strip()
        if s == "":
            raise ValueError("Empty string not allowed")
        s = s.upper()
        return s == s[::-1]
    except Exception as e:
        print("Error:", e)
        return None

try:
    word = input("Enter a string: ")
    result = is_palindrome(word)
    if result is True:
        print("Palindrome")
    elif result is False:
        print("Not Palindrome")
except Exception as e:
    print("Error:", e)
```

Q50. Write a code to generate factorial of the number  For example: factorial of 5 = 5! = 5*4*3*2*1

```python
def factorial(n):
    try:
        if not isinstance(n, int):
            raise ValueError("Input must be an integer")
        if n < 0:
            raise ValueError("Factorial not defined for negative numbers")
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result
    except Exception as e:
        print("Error:", e)
        return None

try:
    num = int(input("Enter a number: "))
    fact = factorial(num)
    if fact is not None:
        print("Factorial =", fact)
except Exception as e:
    print("Error:", e)
```

Q51. Write a program to find largest number in the list.

```python
def find_largest(lst):
    try:
        if not isinstance(lst, list):
            raise ValueError("Input must be a list")
        if len(lst) == 0:
            raise ValueError("List cannot be empty")

        largest = lst[0]
        for num in lst:
            if num > largest:
                largest = num
        return largest

    except Exception as e:
        print("Error:", e)
        return None


try:
    user_input = input("Enter numbers separated by space: ")
    numbers = [float(x) for x in user_input.split()]

    result = find_largest(numbers)

    if result is not None:
        if result.is_integer():
            result = int(result)
        print("Largest number =", result)

except ValueError:
    print("Please enter valid numeric values only.")
except Exception as e:
    print("Error:", e)
```

Q52. Write a program to check frequency of each element in the list.

```python
def frequency_of_elements(lst):
    try:
        if not isinstance(lst, list):
            raise ValueError("Input must be a list")
        if len(lst) == 0:
            raise ValueError("List cannot be empty")

        freq = {}
        for item in lst:
            freq[item] = freq.get(item, 0) + 1
        return freq

    except Exception as e:
        print("Error:", e)
        return None


try:
    user_input = input("Enter elements separated by space: ")
    elements = user_input.split()

    result = frequency_of_elements(elements)

    if result is not None:
        for key, value in result.items():
            print(f"{key} -> {value}")

except Exception as e:
    print("Error:", e)
```

Q53 . There are two string l1 =[ 1,2,3,4,5] and l2 =[3,2,8,7,9] then write a program to find common elements in the list.

```python
def find_common_elements(l1, l2):
    try:
        if not isinstance(l1, list) or not isinstance(l2, list):
            raise ValueError("Both inputs must be lists")

        common = []
        for item in l1:
            if item in l2 and item not in common:
                common.append(item)

        return common

    except Exception as e:
        print("Error:", e)
        return None


try:
    l1 = [1, 2, 3, 4, 5]
    l2 = [3, 2, 8, 7, 9]

    result = find_common_elements(l1, l2)

    if result is not None:
        print("Common elements:", result)

except Exception as e:
    print("Error:", e)
```

```python
import sqlite3
from datetime import datetime

conn = sqlite3.connect("shop.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS user (
    uid INTEGER PRIMARY KEY AUTOINCREMENT,
    mobile TEXT UNIQUE,
    passkey TEXT,
    date TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS product (
    pid INTEGER PRIMARY KEY AUTOINCREMENT,
    pname TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS orders (
    oid INTEGER PRIMARY KEY AUTOINCREMENT,
    uid INTEGER,
    pid INTEGER
)
""")

conn.commit()


class User:
    def __init__(self):
        self.uid = None

    def sign_up(self):
        try:
            mobile = input("Enter mobile: ")
            password = input("Enter password: ")

            if not mobile.isdigit() or len(mobile) != 10:
                raise ValueError("Invalid mobile number")

            cursor.execute(
                "INSERT INTO user (mobile, passkey, date) VALUES (?, ?, ?)",
                (mobile, password, datetime.now())
            )
            conn.commit()
            print("Signup successful")

        except Exception as e:
            print("Error:", e)

    def sign_in(self):
        try:
            mobile = input("Enter mobile: ")
            password = input("Enter password: ")

            cursor.execute(
                "SELECT uid FROM user WHERE mobile=? AND passkey=?",
                (mobile, password)
            )
            result = cursor.fetchone()

            if result:
                self.uid = result[0]
                print("Login successful")
            else:
                print("Invalid credentials")

        except Exception as e:
            print("Error:", e)

    def view_products(self):
        try:
            cursor.execute("SELECT * FROM product")
            products = cursor.fetchall()
            if products:
                for p in products:
                    print(p)
            else:
                print("No products available")
        except Exception as e:
            print("Error:", e)

    def place_order(self):
        try:
            if not self.uid:
                print("Please login first")
                return

            pid = int(input("Enter product id: "))

            cursor.execute("SELECT * FROM product WHERE pid=?", (pid,))
            if cursor.fetchone():
                cursor.execute(
                    "INSERT INTO orders (uid, pid) VALUES (?, ?)",
                    (self.uid, pid)
                )
                conn.commit()
                print("Order placed")
            else:
                print("Invalid product id")

        except Exception as e:
            print("Error:", e)

    def view_orders(self):
        try:
            if not self.uid:
                print("Please login first")
                return

            cursor.execute("""
                SELECT orders.oid, product.pname
                FROM orders
                JOIN product ON orders.pid = product.pid
                WHERE orders.uid=?
            """, (self.uid,))

            orders = cursor.fetchall()
            if orders:
                for o in orders:
                    print(o)
            else:
                print("No orders found")

        except Exception as e:
            print("Error:", e)


class Product:
    def add_product(self):
        try:
            name = input("Enter product name: ")
            if name.strip() == "":
                raise ValueError("Product name cannot be empty")

            cursor.execute("INSERT INTO product (pname) VALUES (?)", (name,))
            conn.commit()
            print("Product added")

        except Exception as e:
            print("Error:", e)


user = User()
product = Product()

while True:
    print("\n1.Sign Up")
    print("2.Sign In")
    print("3.Add Product")
    print("4.View Products")
    print("5.Place Order")
    print("6.View Orders")
    print("7.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        user.sign_up()
    elif choice == "2":
        user.sign_in()
    elif choice == "3":
        product.add_product()
    elif choice == "4":
        user.view_products()
    elif choice == "5":
        user.place_order()
    elif choice == "6":
        user.view_orders()
    elif choice == "7":
        print("Exiting...")
        break
    else:
        print("Invalid choice")

conn.close()
```
