Using Dictionary

```python
hotel_name = "Welcome to Shrinath's Cafe"


menu = {
    1: ("idli", 30),
    2: ("dosa", 50),
    3: ("poha", 25),
    4: ("upma", 20)
}


print("--------- MENU ---------")
for sr, (item, price) in menu.items():
    print(f"{sr}. {item} - ₹{price}")
print("------------------------")


choice = None
while choice is None:
    try:
        choice_input = input("Enter the sr number of the item you want: ")
        choice = int(choice_input)
        if choice not in menu:
            print("Invalid serial number. Please choose from the menu.")
            choice = None
    except ValueError:
        print("Invalid input. Please enter a number corresponding to the item.")
        choice = None

quantity = None
while quantity is None:
    try:
        quantity_input = input("Enter quantity: ")
        quantity = int(quantity_input)
        if quantity <= 0:
            print("Quantity must be a positive number.")
            quantity = None
    except ValueError:
        print("Invalid quantity. Please enter a number.")
        quantity = None



item_name, item_price = menu[choice]


items = [(choice, item_name, quantity, item_price)]


total = sum(qty * price for _, _, qty, price in items)


print("=" * 50)
print(f"|{hotel_name.center(48)}|")
print("=" * 50)
print(f"| {'sr':<3} {'Menu':<15} {'qunt':<8} {'price':>17} |")
print("-" * 50)

for sr, menu_item, qty, price in items:
    print(f"| {sr:<3} {menu_item:<15} {qty:<8} {price:>17} |")

print("-" * 50)
print(f"| {'Total':<30} {total:>15} |")
print("=" * 50)
```

Function version of Bill Printing

```python
def hotel_bill_printing():

    hotel_name = "Welcome to Shrinath's Cafe"

    menu = {
        1: ("idli", 30),
        2: ("dosa", 50),
        3: ("poha", 25),
        4: ("upma", 20)
    }

    print("--------- MENU ---------")
    for sr, (item, price) in menu.items():
        print(f"{sr}. {item} - ₹{price}")
    print("------------------------")

    while True:
        try:
            choice = int(input("Enter the sr number of the item you want: "))
            if choice in menu:
                break
            else:
                print("Invalid serial number. Please choose from the menu.")
        except ValueError:
            print("Please enter a valid number.")

    while True:
        try:
            quantity = int(input("Enter quantity: "))
            if quantity > 0:
                break
            else:
                print("Quantity must be positive.")
        except ValueError:
            print("Please enter a valid number.")

    item_name, item_price = menu[choice]
    total = quantity * item_price

    print("=" * 50)
    print(f"|{hotel_name.center(48)}|")
    print("=" * 50)
    print(f"| {'sr':<3} {'Menu':<15} {'Qty':<8} {'Price':>17} |")
    print("-" * 50)
    print(f"| {choice:<3} {item_name:<15} {quantity:<8} {item_price:>17} |")
    print("-" * 50)
    print(f"| {'Total':<30} {total:>15} |")
    print("=" * 50)


hotel_bill_printing()
```

OOPs version of Bill Printing

```python
class HotelBill:

    def __init__(self):
        self.hotel_name = "Welcome to Shrinath's Cafe"
        self.menu = {
            1: ("idli", 30),
            2: ("dosa", 50),
            3: ("poha", 25),
            4: ("upma", 20)
        }
        self.items = []

    def display_menu(self):
        print("--------- MENU ---------")
        for sr, (item, price) in self.menu.items():
            print(f"{sr}. {item} - ₹{price}")
        print("------------------------")

    def take_order(self):
        while True:
            try:
                choice = int(input("Enter item sr number (0 to stop): "))

                if choice == 0:
                    break

                if choice not in self.menu:
                    print("Invalid serial number.")
                    continue

                quantity = int(input("Enter quantity: "))
                if quantity <= 0:
                    print("Quantity must be positive.")
                    continue

                item_name, item_price = self.menu[choice]
                self.items.append((choice, item_name, quantity, item_price))

            except ValueError:
                print("Please enter valid numbers only.")

    def print_bill(self):
        if not self.items:
            print("No items selected.")
            return

        total = sum(qty * price for _, _, qty, price in self.items)

        print("=" * 50)
        print(f"|{self.hotel_name.center(48)}|")
        print("=" * 50)
        print(f"| {'sr':<3} {'Menu':<15} {'Qty':<8} {'Price':>17} |")
        print("-" * 50)

        for sr, menu_item, qty, price in self.items:
            print(f"| {sr:<3} {menu_item:<15} {qty:<8} {qty * price:>17} |")

        print("-" * 50)
        print(f"| {'Total':<30} {total:>15} |")
        print("=" * 50)


bill = HotelBill()
bill.display_menu()
bill.take_order()
bill.print_bill()
```
