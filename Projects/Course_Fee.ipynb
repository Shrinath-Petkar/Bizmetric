 Ask user to opt for courses for master degree based on the following

L1 = [“HR”, “Finance”, “Marketing”, “DS”]

Based on above subject there are two different streams. For example- HR is having HR core and HR analytics and Marketing is having core and Marketing analytics. Analytics is the optional subject and having added extra fees. DS is not having analytics.

If fees for L1 is 2 lakhs for each course core subject having the same fees but analytics subject having 10% extra on 2 lakhs.

If student opts for hostel then 2 lakhs per year is added. For food monthly 2000 .

Transportation charges 13000 per semester. Calculate the total annual cost based on selected service.

User will enter values as subject, analytics(Y/N), Hostel (Y/N), food(How many months?), Transportation(semester/annual)

Normal if-else version of Course Fee

```python
L1 = ["HR", "Finance", "Marketing", "DS"]

BASE_FEE = 200000
ANALYTICS_EXTRA = 0.10
HOSTEL_FEE = 200000
FOOD_PER_MONTH = 2000
TRANSPORT_PER_SEM = 13000


subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
analytics = input("Analytics? (Y/N): ").strip().upper()
hostel = input("Hostel? (Y/N): ").strip().upper()
food_months = int(input("Food for how many months?: "))
transport = input("Transportation (semester/annual): ").strip().lower()


if subject not in L1:
    print("Invalid subject selected!")
    exit()


total_fee = BASE_FEE


if analytics == "Y" and subject != "DS":
    total_fee += BASE_FEE * ANALYTICS_EXTRA


if hostel == "Y":
    total_fee += HOSTEL_FEE

total_fee += food_months * FOOD_PER_MONTH

if transport == "semester":
    total_fee += TRANSPORT_PER_SEM * 2
elif transport == "annual":
    total_fee += TRANSPORT_PER_SEM * 2
else:
    print("Invalid transportation option!")
    exit()

print("\n----- Fee Breakdown -----")
print(f"Subject: {subject}")
print(f"Total Annual Cost: ₹{int(total_fee)}")
```

Function version of Course Fee

```python
L1 = ["HR", "Finance", "Marketing", "DS"]

BASE_FEE = 200000
ANALYTICS_EXTRA = 0.10
HOSTEL_FEE = 200000
FOOD_PER_MONTH = 2000
TRANSPORT_PER_SEM = 13000

def calculate_total_fee():

    while True:
        subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
        if subject in L1:
            break
        print("Invalid subject. Please choose from HR, Finance, Marketing, DS.")

    while True:
        analytics = input("Analytics? (Y/N): ").strip().upper()
        if analytics in ["Y", "N"]:
            break
        print("Enter Y or N only.")

    while True:
        hostel = input("Hostel? (Y/N): ").strip().upper()
        if hostel in ["Y", "N"]:
            break
        print("Enter Y or N only.")

    while True:
        try:
            food_months = int(input("Food for how many months?: "))
            if food_months >= 0:
                break
            print("Months cannot be negative.")
        except ValueError:
            print("Please enter a valid number.")

    while True:
        transport = input("Transportation (semester/annual): ").strip().lower()
        if transport in ["semester", "annual"]:
            break
        print("Enter 'semester' or 'annual' only.")

    total_fee = BASE_FEE

    if analytics == "Y" and subject != "DS":
        total_fee += BASE_FEE * ANALYTICS_EXTRA

    if hostel == "Y":
        total_fee += HOSTEL_FEE

    total_fee += food_months * FOOD_PER_MONTH

    total_fee += TRANSPORT_PER_SEM * 2

    print("\n----- Fee Breakdown -----")
    print(f"Subject: {subject}")
    print(f"Total Annual Cost: ₹{int(total_fee)}")


calculate_total_fee()
```

```python
class FeeCalculator:

    L1 = ["HR", "Finance", "Marketing", "DS"]

    BASE_FEE = 200000
    ANALYTICS_EXTRA = 0.10
    HOSTEL_FEE = 200000
    FOOD_PER_MONTH = 2000
    TRANSPORT_PER_SEM = 13000

    def __init__(self, subject, analytics, hostel, food_months, transport):
        self.subject = subject
        self.analytics = analytics
        self.hostel = hostel
        self.food_months = food_months
        self.transport = transport
        self.total_fee = 0

    def calculate_fee(self):

        self.total_fee = self.BASE_FEE

        if self.analytics == "Y" and self.subject != "DS":
            self.total_fee += self.BASE_FEE * self.ANALYTICS_EXTRA

        if self.hostel == "Y":
            self.total_fee += self.HOSTEL_FEE

        self.total_fee += self.food_months * self.FOOD_PER_MONTH

        if self.transport in ["semester", "annual"]:
            self.total_fee += self.TRANSPORT_PER_SEM * 2

    def display(self):
        print("\n----- Fee Breakdown -----")
        print(f"Subject: {self.subject}")
        print(f"Total Annual Cost: ₹{int(self.total_fee)}")



while True:
    subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
    if subject in FeeCalculator.L1:
        break
    print("Invalid subject.")

while True:
    analytics = input("Analytics? (Y/N): ").strip().upper()
    if analytics in ["Y", "N"]:
        break
    print("Enter Y or N only.")

while True:
    hostel = input("Hostel? (Y/N): ").strip().upper()
    if hostel in ["Y", "N"]:
        break
    print("Enter Y or N only.")

while True:
    try:
        food_months = int(input("Food for how many months?: "))
        if food_months >= 0:
            break
        print("Months cannot be negative.")
    except ValueError:
        print("Enter valid number.")

while True:
    transport = input("Transportation (semester/annual): ").strip().lower()
    if transport in ["semester", "annual"]:
        break
    print("Enter 'semester' or 'annual' only.")


student = FeeCalculator(subject, analytics, hostel, food_months, transport)
student.calculate_fee()
student.display()
```

SQL version of Course Fee

```python
import sqlite3

class FeeCalculator:

    L1 = ["HR", "Finance", "Marketing", "DS"]

    BASE_FEE = 200000
    ANALYTICS_EXTRA = 0.10
    HOSTEL_FEE = 200000
    FOOD_PER_MONTH = 2000
    TRANSPORT_PER_SEM = 13000

    def __init__(self, subject, analytics, hostel, food_months, transport):
        self.subject = subject
        self.analytics = analytics
        self.hostel = hostel
        self.food_months = food_months
        self.transport = transport
        self.total_fee = 0

    def calculate_fee(self):

        self.total_fee = self.BASE_FEE

        if self.analytics == "Y" and self.subject != "DS":
            self.total_fee += self.BASE_FEE * self.ANALYTICS_EXTRA

        if self.hostel == "Y":
            self.total_fee += self.HOSTEL_FEE

        self.total_fee += self.food_months * self.FOOD_PER_MONTH

        self.total_fee += self.TRANSPORT_PER_SEM * 2

    def save_to_db(self, cursor):
        cursor.execute("""
        INSERT INTO student_fee
        (subject, analytics, hostel, food_months, transport, total_fee)
        VALUES (?, ?, ?, ?, ?, ?)
        """, (self.subject, self.analytics, self.hostel,
              self.food_months, self.transport, self.total_fee))

    def display(self):
        print("\n----- Fee Breakdown -----")
        print(f"Subject: {self.subject}")
        print(f"Total Annual Cost: ₹{int(self.total_fee)}")



conn = sqlite3.connect("fees.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS student_fee (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    subject TEXT,
    analytics TEXT,
    hostel TEXT,
    food_months INTEGER,
    transport TEXT,
    total_fee REAL
)
""")

while True:
    print("\n1. Add Student Fee")
    print("2. Display Records")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        while True:
            subject = input("Enter subject (HR/Finance/Marketing/DS): ").strip()
            if subject in FeeCalculator.L1:
                break
            print("Invalid subject.")

        analytics = input("Analytics? (Y/N): ").strip().upper()
        hostel = input("Hostel? (Y/N): ").strip().upper()
        food_months = int(input("Food months: "))
        transport = input("Transportation (semester/annual): ").strip().lower()

        student = FeeCalculator(subject, analytics, hostel, food_months, transport)
        student.calculate_fee()
        student.save_to_db(cursor)
        conn.commit()

        student.display()
        print("Record saved successfully!")

    elif choice == "2":
        cursor.execute("SELECT * FROM student_fee")
        records = cursor.fetchall()

        if records:
            print("\n--- Student Fee Records ---")
            for row in records:
                print(f"ID:{row[0]} | Subject:{row[1]} | "
                      f"Total Fee: ₹{int(row[6])}")
        else:
            print("No records found.")

    elif choice == "3":
        break

    else:
        print("Invalid choice.")

conn.close()
```
